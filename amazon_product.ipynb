```python
# Importing some vital lib.
from selenium import webdriver
import time #standard lib
```

```python
#creating a class which is responsible for search and extracting prices of items
class AmazonBot(object): # calling to the object class.
    def __init__(self,items): #item is list of items extracted from googlesheet
        self.amazon_url = 'https://www.amazon.in/'
        self.items = items
        
        self.driver = webdriver.Chrome() # object of the chrome(browser)
        # You can use firefox and Internet explorer but you have to use specific methods or funtions.
        # note: keep an eye on browers version.
    
    def search_items(self):
        urls = []
        prices=[]
        names=[]
        print(f'Items is {self.items}')
        
        for item in self.items:
            
            try:
                print(f'searching for {item}.')
                self.driver.get(self.amazon_url) # opening the  Amazon webpage.
                time.sleep(4) # stoping interpter about 4sec.
                # find the tag whose id is 'twotabssearchtextbox'.
                search_input_bar = self.driver.find_element_by_id('twotabsearchtextbox')
                time.sleep(4) # stoping interpter about 4sec.

                # send the string to search bar.
                search_input_bar.send_keys(item)
                print('sending the keys to search bar')
                time.sleep(4)
                
                 # find the tag which have class_name as 'nav-input'
                search_button = self.driver.find_element_by_class_name('nav-input')

                time.sleep(8) # sleep for 8 sec

                search_button.click() # clicking on the search.
                print('Click to start the search.')
                time.sleep(8)
                
                # search the tag which consist of xpath.
                res = self.driver.find_element_by_xpath('//*[@id="search"]/div[1]/div/div[1]/div/span[3]/div[2]/div[2]')
                print('found the desired xpath.')

                time.sleep(8)
                
                # extracting a number from 'res'.
                link = res.get_attribute('data-asin')
                print('link is available')
                
                # concatenation the 2 strings.
                url = 'https://amazon.in/dp/' + link

                price = self.get_product_price(url)

                name = self.get_product_name(url)
                print('***********************')
                print(name)
                print('***********************')
                print(price)
                print('***********************')
                print(url)
                
                
                # adding the results to lists.
                urls.append(url)
                prices.append(price)
                names.append(name)

            
            
            # if anyerror occurs exception will handle it.
            except:
                print('Your Internet is slow or Unable to find the specific tag')
                
                self.driver.close() # close the browser.
            

            
    def get_product_price(self, url):
        
        self.driver.get(url) # get the url
        
        try:
            # find the tag with id of 'priceblock_ourprice' and from there I'll extract the text
            price = self.driver.find_element_by_id('priceblock_ourprice').text
        
        except:
            pass 
        if price == None:
            prive = 'product is unavailable for at the moment.'
        else:
            pass
        
        return price
    
    def get_product_name(self, url):
        
        self.driver.get(url)
        
        try:
            product_name = self.driver.find_element_by_id('productTitle').text
            
        except:
            pass
        
        if product_name == None:
            print('product is unavaliable at the moment')
        
        else:
            pass
        
        return product_name


ab = AmazonBot(['redmi note 9'])            
ab.search_items()
```
